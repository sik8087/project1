```
from pyspark.ml import *
from pyspark.sql.functions import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyspark.sql.functions import mean as _mean, stddev as _stddev, col
from pyspark.sql.functions import when
```

```
df1 = spark.read.csv("/FileStore/tables/OOT_Bureau.csv",header=True,inferSchema=True)
```

```
df1.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df1.columns]).show()
```

```
lst_str = []
lst_str = [item[0] for item in df1.dtypes if item[1].startswith('string')]
lst_str
```

```
for i in df1.columns:
    df1.select(i).filter(df1[i]<-10).groupby(i).count().show()
    
```

```
for i in df1.columns:
    if i not in lst_str:
        df1 = df1.withColumn(i,when(col(i)==  -9999,0).otherwise(col(i)))
        df1 = df1.withColumn(i,when(col(i)==  -11111,0).otherwise(col(i)))
        df1 = df1.withColumn(i,when(col(i)<= -10,df1.groupby(i).count().orderBy("count", ascending=False).first()[0]).otherwise(col(i)))
```

```
for i in df1.columns:
    df1.select(i).filter(df1[i]<-10).groupby(i).count().show()
```

```
for i in df1.columns:
    if i not in lst_str:
        df1 = df1.withColumn(i,when(col(i)<=  -10,0).otherwise(col(i)))
```
